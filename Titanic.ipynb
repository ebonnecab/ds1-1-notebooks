```python
import pandas
import csv
```

```python
with open('titanic.csv') as csv_file:
    reader = csv_file.readlines()
```

```python
data = [row.split(',') for row in reader]
```

```python
for item in data:
    female_survivors = []
    if item[1] == '1' and item[5] == 'female':
        female_survivors.append(item)
    for person in female_survivors:
        survival_count = 0
        if person[1] == '1':
            survival_count += 1
            print(survival_count)
    
```
